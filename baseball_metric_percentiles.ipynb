# ***The following code is used to analyze a hitter's Swing and Take Decisions (SDS)***

```python
from scipy.stats import norm
from sklearn.model_selection import train_test_split
import matplotlib as plt
from sklearn import datasets
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
import joblib
from pandas import DataFrame
import statistics
from scipy.stats import norm

df3 = batdf.copy()

contactProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/ContactProbModel.joblib')
swingProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/SwingProbModel.joblib')
strikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/StrikeProbModel.joblib')
hhbProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/HardHitProbModel.joblib')
```

```python
# HOLDS FUNCTIONS THAT CALCUALTE THE SDS+ OF EACH ROW

# function to calculate the Swing SDS+ of each player
def swingSDS(filter_swing: DataFrame) -> DataFrame:
  # import everything
  import joblib
  # contactProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/ContactProbModel.joblib')
  # swingProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/SwingProbModel.joblib')
  # strikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/StrikeProbModel.joblib')
  # hhbProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/HardHitProbModel.joblib')

  # makes dataframes that apply models
  cleanedSwing = filter_swing[['RelSpeed', 'SpinRate', 'InducedVertBreak', 'HorzBreak', 'PlateLocHeight', 'PlateLocSide', 'VertApprAngle', 'HorzApprAngle']]
  strikeSwing = filter_swing[['PlateLocHeight', 'PlateLocSide']]

  # applies models to datasets
    # swings
  swingContact = contactProb.predict_proba(cleanedSwing)
  swingSwing = swingProb.predict_proba(cleanedSwing)
  swingStrike = strikeProb.predict_proba(strikeSwing)
  swingHHB = hhbProb.predict_proba(cleanedSwing)

  # variables for swings
  swingSDS = 0.0
  swingContactProb = 0.0
  swingSwingProb = 0.0
  swingStrikeProb = 0.0
  swingHHBProb = 0.0

  # lists to hold values
  swingSDS_list = []
  swingContactProb_list = []
  swingSwingProb_list = []
  swingStrikeProb_list = []
  swingHHBProb_list = []

  # probability lists
  swingPContact_list = []
  swingPHHB_list = []

  # swing SDS Information
  swingSDSMean = 1.8926646183979023
  swingSDSRange = 4.0
  swingSDSSTD = 0.9558850380546416
  swingSDSMin = -1.0
  swingSDSMax = 3.0

  # probablities and SDS
  for i in range(len(cleanedSwing)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    swingContactProb = swingContact[i][1]
    swingSwingProb = swingSwing[i][1]
    swingStrikeProb = swingStrike[i][1]
    swingHHBProb = swingHHB[i][1]

    # adds probabilites to probability list
    swingContactProb_list.append(swingContactProb)
    swingSwingProb_list.append(swingSwingProb)
    swingStrikeProb_list.append(swingStrikeProb)
    swingHHBProb_list.append(swingHHBProb)

    # calculate and store probabilites used for SDS
    swingPContact_list.append( (swingContactProb*swingSwingProb) / swingSwingProb )
    swingPHHB_list.append( (swingHHBProb*swingPContact_list[i]) / swingPContact_list[i] )

    # calculate and store SDS Swing
    swingSDS = (swingStrikeProb_list[i] - (1-swingStrikeProb_list[i])) + swingSwingProb_list[i] + swingHHBProb_list[i]
    swingSDS_list.append(( ( ( (swingSDS - swingSDSMin) - (swingSDSMean - swingSDSMin) )/swingSDSSTD)*100) + 100)
    #swingSDS_list.append(swingSDS)

  filter_swing['SwingSDS'] = swingSDS_list

  # returns the dataframe that now consists of Swing SDS+
  return filter_swing


# function to calculate the Take SDS+ of each row
def takeSDS(filter_take: DataFrame) -> DataFrame:
  # import everything
  import joblib
  # contactProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/ContactProbModel.joblib')
  # swingProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/SwingProbModel.joblib')
  # strikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/StrikeProbModel.joblib')
  # hhbProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/HardHitProbModel.joblib')

  # makes dataframes that apply models
  cleanedTake = filter_take[['RelSpeed', 'SpinRate', 'InducedVertBreak', 'HorzBreak', 'PlateLocHeight', 'PlateLocSide', 'VertApprAngle', 'HorzApprAngle']]
  strikeTake = filter_take[['PlateLocHeight', 'PlateLocSide']]

  # applies models to datasets
    # takes
  takeContact = contactProb.predict_proba(cleanedTake)
  takeSwing = swingProb.predict_proba(cleanedTake)
  takeStrike = strikeProb.predict_proba(strikeTake)
  takeHHB = hhbProb.predict_proba(cleanedTake)

  # variables for takes
  takeSDS = 0.0
  takeContactProb = 0.0
  takeSwingProb = 0.0
  takeStrikeProb = 0.0
  takeHHBProb = 0.0

  # lists to hold values
  takeSDS_list = []
  takeContactProb_list = []
  takeSwingProb_list = []
  takeStrikeProb_list = []
  takeHHBProb_list = []

  # probability lists
  takePContact_list = []
  takePHHB_list = []

   # take SDS Information
  takeSDSMean = -0.07691706404188041
  takeSDSRange = 3.966666666666667
  takeSDSSTD = 1.1273762539917347
  takeSDSMin = -2.966666666666667
  takeSDSMax = 1.0

  # probablities and SDS
  for i in range(len(cleanedTake)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    takeContactProb = takeContact[i][1]
    takeSwingProb = takeSwing[i][1]
    takeStrikeProb = takeStrike[i][1]
    takeHHBProb = takeHHB[i][1]

    # adds probabilites to probability list
    takeContactProb_list.append(takeContactProb)
    takeSwingProb_list.append(takeSwingProb)
    takeStrikeProb_list.append(takeStrikeProb)
    takeHHBProb_list.append(takeHHBProb)

    # calculate and store probabilites used for SDS
    takePContact_list.append( (takeContactProb*takeSwingProb) / takeSwingProb )
    takePHHB_list.append( (takeHHBProb*takePContact_list[i]) / takePContact_list[i] )

    # calculate and store SDS Take
    takeSDS = ( (1-takeStrikeProb_list[i]) - takeStrikeProb_list[i] ) - takeSwingProb_list[i] - takeHHBProb_list[i]
    takeSDS_list.append(( ( ( (takeSDS - takeSDSMin) - (takeSDSMean - takeSDSMin) )/takeSDSSTD)*100) + 100)


  filter_take['TakeSDS'] = takeSDS_list

  # returns the dataframe that now consists of Take SDS+
  return filter_take
```

# ***The following code is used to calculate/measure a pitcher's Location+***

```python
# HOLDS FUNCTIONS THAT CALCUALTE THE LOCATION+ OF EACH PITCH

# function to calculate the Location+ of a pitcher's fastball
def fb_location(fb: DataFrame) -> DataFrame:
  # retains important columns
  cleanedFB = fb[['PlateLocHeight', 'PlateLocSide']]

  # imports models
  fbStrikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/FBStrikeProbModel.joblib')
  fbHardHitProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/FBHardHitProbModel.joblib')

  # applies models to datasets
  fbStrike = fbStrikeProb.predict_proba(cleanedFB) # Strike probability model
  fbHardHit = fbHardHitProb.predict_proba(cleanedFB) # Hard hit probability model

  # variables for FBs
  fbLocation = 0.0
  fbStrikeProb = 0.0
  fbHHBProb = 0.0

  # lists to hold values
  fbLocation_list = []
  fbStrikeProb_list = []
  fbHHBProb_list = []

  # Fastball Location+ Information
  fbLocation_Mean = 0.406058499550797
  fbLocation_STD = 0.3532781763890214
  fbLocation_Min = -0.45

  # probabilitying and Location+
  for i in range(len(cleanedFB)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    fbStrikeProb = fbStrike[i][1]
    fbHHBProb = fbHardHit[i][1]

    # adds probabilites to probability list
    fbStrikeProb_list.append(fbStrikeProb)
    fbHHBProb_list.append(fbHHBProb)

    # calculate and store Location Rating
    fbLocation = (fbStrikeProb_list[i] - 0.45*(fbHHBProb_list[i]))
    fbLocation_list.append(( ( ( (fbLocation - fbLocation_Min) - (fbLocation_Mean - fbLocation_Min) )/fbLocation_STD)*100) + 100)

  # adds Location column to dataframe
  fb['Location+'] = fbLocation_list

  # returns the Location+ value
  return fb



# function to calculate the Location+ of a pitcher's slider
def sl_location(sl: DataFrame) -> DataFrame:
  # retain important columns
  cleanedSL = sl[['PlateLocHeight', 'PlateLocSide']]

  # imports models
  slStrikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/SLStrikeProbModel.joblib')
  slHardHitProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/SLHardHitProbModel.joblib')

  # applies models to datasets
  slStrike = slStrikeProb.predict_proba(cleanedSL) # Strike probability model
  slHardHit = slHardHitProb.predict_proba(cleanedSL) # Hard hit probability model

  # variables for SLs
  slLocation = 0.0
  slStrikeProb = 0.0
  slHHBProb = 0.0

  # lists to hold values
  slLocation_list = []
  slStrikeProb_list = []
  slHHBProb_list = []

  # Slider Location+ Information
  slLocation_Mean = 0.34960684689716576
  slLocation_STD = 0.3533850412263207
  slLocation_Min = -0.45

  # probabilitying and Location+
  for i in range(len(cleanedSL)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    slStrikeProb = slStrike[i][1]
    slHHBProb = slHardHit[i][1]

    # adds probabilites to probability list
    slStrikeProb_list.append(slStrikeProb)
    slHHBProb_list.append(slHHBProb)

    # calculate and store Location Rating
    slLocation = (slStrikeProb_list[i] - 0.45*(slHHBProb_list[i]))
    slLocation_list.append((((slLocation - slLocation_Min) - (slLocation_Mean - slLocation_Min)) / slLocation_STD) * 100 + 100)

  # adds Location column to dataframe
  sl['Location+'] = slLocation_list

  # returns the Location+ value
  return sl



# function to calculate the Location+ of a pitcher's changeup
def ch_location(ch: DataFrame) -> DataFrame:
  # retain important columns
  cleanedCH = ch[['PlateLocHeight', 'PlateLocSide']]

  # imports models
  chStrikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/CHStrikeProbModel.joblib')
  chHardHitProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/CHHardHitProbModel.joblib')

  # applies models to datasets
  chStrike = chStrikeProb.predict_proba(cleanedCH) # Strike probability model
  chHardHit = chHardHitProb.predict_proba(cleanedCH) # Hard hit probability model

  # variables for CHs
  chLocation = 0.0
  chStrikeProb = 0.0
  chHHBProb = 0.0

  # lists to hold values
  chLocation_list = []
  chStrikeProb_list = []
  chHHBProb_list = []

  # ChangeUp Location+ Information
  chLocation_Mean = 0.3353856241831144
  chLocation_STD = 0.3451946132850606
  chLocation_Min = -0.45

  # probabilitying and Location+
  for i in range(len(cleanedCH)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    chStrikeProb = chStrike[i][1]
    chHHBProb = chHardHit[i][1]

    # adds probabilites to probability list
    chStrikeProb_list.append(chStrikeProb)
    chHHBProb_list.append(chHHBProb)

    # calculate and store Location Rating
    chLocation = (chStrikeProb_list[i] - 0.45*(chHHBProb_list[i]))
    chLocation_list.append((((chLocation - chLocation_Min) - (chLocation_Mean - chLocation_Min)) / chLocation_STD) * 100 + 100)

  # adds Location column to dataframe
  ch['Location+'] = chLocation_list

  # returns the Location+ value
  return ch




# function to calculate the Location+ of a pitcher's curveball
def cu_location(cu: DataFrame) -> DataFrame:
  # retain important columns
  cleanedCU = cu[['PlateLocHeight', 'PlateLocSide']]

   # imports models
  cuStrikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/CUStrikeProbModel.joblib')
  cuHardHitProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/CUHardHitProbModel.joblib')

  # applies models to datasets
  cuStrike = cuStrikeProb.predict_proba(cleanedCU) # Strike probability model
  cuHardHit = cuHardHitProb.predict_proba(cleanedCU) # Hard hit probability model

  # variables for CUs
  cuLocation = 0.0
  cuStrikeProb = 0.0
  cuHHBProb = 0.0

  # lists to hold values
  cuLocation_list = []
  cuStrikeProb_list = []
  cuHHBProb_list = []

  # Curveball Location+ Information
  cuLocation_Mean = 0.3088163519897403
  cuLocation_STD = 0.35590971460457
  cuLocation_Min = -0.45

  # probabilitying and Location+
  for i in range(len(cleanedCU)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    cuStrikeProb = cuStrike[i][1]
    cuHHBProb = cuHardHit[i][1]

    # adds probabilites to probability list
    cuStrikeProb_list.append(cuStrikeProb)
    cuHHBProb_list.append(cuHHBProb)

    # calculate and store Location Rating
    cuLocation = (cuStrikeProb_list[i] - 0.45*(cuHHBProb_list[i]))
    cuLocation_list.append((((cuLocation - cuLocation_Min) - (cuLocation_Mean - cuLocation_Min)) / cuLocation_STD) * 100 + 100)

  # adds Location column to dataframe
  cu['Location+'] = cuLocation_list

  # returns the Location+ value
  return cu




# function to calculate the Location+ of a pitcher's other pitches
def sinker_location(sinker: DataFrame) -> DataFrame:
  # retain important columns
  cleanedSinker = sinker[['PlateLocHeight', 'PlateLocSide']]

   # imports models
  sinkerStrikeProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/FBStrikeProbModel.joblib')
  sinkerHardHitProb = joblib.load('/content/drive/Shareddrives/USC Baseball/SabermetricModels/FBHardHitProbModel.joblib')

  # applies models to datasets
  sinkerStrike = sinkerStrikeProb.predict_proba(cleanedSinker) # Strike probability model
  sinkerHardHit = sinkerHardHitProb.predict_proba(cleanedSinker) # Hard hit probability model

  # variables for Sinkers
  sinkerLocation = 0.0
  sinkerStrikeProb = 0.0
  sinkerHHBProb = 0.0

  # lists to hold values
  sinkerLocation_list = []
  sinkerStrikeProb_list = []
  sinkerHHBProb_list = []

  # Sinker Location+ Information
  sinkerLocation_Mean = 0.406058499550797
  sinkerLocation_STD = 0.3532781763890214
  sinkerLocation_Min = -0.45

  # probabilitying and Location+
  for i in range(len(cleanedSinker)):
    # predicts the probability of a pitch being contact, swing, strike, and HHB
    sinkerStrikeProb = sinkerStrike[i][1]
    sinkerHHBProb = sinkerHardHit[i][1]

    # adds probabilites to probability list
    sinkerStrikeProb_list.append(sinkerStrikeProb)
    sinkerHHBProb_list.append(sinkerHHBProb)

    # calculate and store Location Rating
    sinkerLocation = (sinkerStrikeProb_list[i] - 0.45*(sinkerHHBProb_list[i]))
    sinkerLocation_list.append((((sinkerLocation - sinkerLocation_Min) - (sinkerLocation_Mean - sinkerLocation_Min)) / sinkerLocation_STD) * 100 + 100)

  # adds location to the dataframe
  sinker['Location+'] = sinkerLocation_list

  # returns the Location+ value
  return sinker
```
